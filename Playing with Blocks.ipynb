```python
# Make a new 20 by 20 block grid. Use a `for` loop with `range` to change the
# color of every block in the third, 10th, and 18th rows.
for column_index in range(grid.width):
    grid[2, column_index].rgb = colors.CadetBlue
    grid.flash()
    grid[9, column_index].rgb = colors.CadetBlue
    grid.flash()
    grid[17, column_index].rgb = colors.CadetBlue
    grid.flash()
grid
```

```python
# Make a new 20 by 20 block grid. Use a `for` loop with `range` to change the
# color of every block in the third, 10th, and 18th rows.
for column_index in range(grid.width):
    for row_index in [2, 9, 17]:
        grid[row_index, column_index].rgb = colors.Gray
        grid.flash()
```

```python
# Use nested `for` loops with `range` to change the color of the bottom-left
# quadrant of your 20 by 20 grid.
for row_index in range(int(grid.height / 2), grid.height):
    for column_index in range(grid.width // 2):
        grid[row_index, column_index].rgb = colors.Navy
        grid.flash()
```

```python
# Write a function that takes a grid as input and returns a new grid with
# the colors inverted (e.g. white becomes black, black becomes white,
# yellow becomes ???).
def invert_grid(grid):
    # new_grid = BlockGrid(grid.width, grid.height)
    new_grid = grid.copy()
    for row_index in range(grid.height):
        for column_index in range(grid.width):
            new_grid[row_index, column_index].red = 255 - grid[row_index, column_index].red
            new_grid[row_index, column_index].green = 255 - grid[row_index, column_index].green
            new_grid[row_index, column_index].blue = 255 - grid[row_index, column_index].blue
    return new_grid
```

```python
# Write a function that takes a grid as input and returns a new grid with
# the colors inverted (e.g. white becomes black, black becomes white,
# yellow becomes ???).

def invert_color(color_value):
    return 255 - color_value

def invert_grid2(grid):
    # new_grid = BlockGrid(grid.width, grid.height)
    new_grid = grid.copy()
    for (new_block, old_block) in zip(new_grid, grid):
        new_block.rgb = [invert_color(color_value) for color_value in old_block.rgb]
    return new_grid
```
